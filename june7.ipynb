```python
df=pd.read_csv('data_cleaned.csv')
df
```

```python
x=df.drop(['Survived'],axis=1)
y=df['Survived']

from sklearn.preprocessing import MinMaxScaler 
scaled=MinMaxScaler()
scaled_x=scaled.fit_transform(x)
x=pd.DataFrame(scaled_x,columns=x.columns)
from sklearn.model_selection import train_test_split as tt
train_x,val_x,train_y,val_y=tt(x,y,random_state=101,stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression 
lr=LogisticRegression()
lr.fit(train_x,train_y)
lr.score(train_x,train_y)
```

```python
predict1=lr.predict(val_x)
lr.score(val_x,val_y),predict1
```

```python
from sklearn.neighbors import KNeighborsClassifier 
knn=KNeighborsClassifier(n_neighbors=5)

knn.fit(train_x,train_y)
knn.score(train_x,train_y)
```

```python
predict2=lr.predict(val_x)
knn.score(val_x,val_y),predict2
```

```python
from sklearn.tree import DecisionTreeClassifier 
dd=DecisionTreeClassifier(max_depth=10)
dd.fit(train_x,train_y)
dd.score(train_x,train_y)
```

```python
predict3=lr.predict(val_x)
dd.score(val_x,val_y),predict3
```

```python
from statistics import mode
final1=np.array([])
for i in range(len(val_x)):
    final1=np.append(final1,mode([predict1[i],predict2[i],predict3[i]]))
final1
```

```python
from sklearn.metrics import accuracy_score as af
```

```python
af(val_y,final1)
```

```python
df=pd.read_csv('train_cleaned.csv')
df
```

```python
x=df.drop(['Item_Outlet_Sales'],axis=1)
y=df['Item_Outlet_Sales']
```

```python
from sklearn.model_selection import train_test_split as tt
train_x,val_x,train_y,val_y=tt(x,y,random_state=101,shuffle=False)
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
ll=LinearRegression()
ll.fit(train_x,train_y)
ll.score(train_x,train_y)
prediction1=ll.predict(val_x)
prediction1
a=ll.score(val_x,val_y)
a
```

```python
knn=KNeighborsRegressor(n_neighbors=9)
knn.fit(train_x,train_y)
knn.score(train_x,train_y)
prediction2=knn.predict(val_x)
prediction2
b=knn.score(val_x,val_y)
b
```

```python
dd=DecisionTreeRegressor(max_depth=7)
dd.fit(train_x,train_y)
dd.score(train_x,train_y)
prediction3=dd.predict(val_x)
prediction3
c=dd.score(val_x,val_y)
c
```

```python
from statistics import mean
final2=np.array([])
for i in range(len(val_x)):
    final2=np.append(final2,mean([prediction1[i],prediction2[i],prediction3[i]]))
    

```

```python
from sklearn.metrics import r2_score as rr
rr(val_y,final2)
```

```python
from statistics import mean
final2=np.array([])
for i in range(len(val_x)):
    final2=np.append(final2,mean([prediction1[i],prediction1[i],prediction2[i],prediction3[i],prediction3[i]]))
```

```python
rr(val_y,final2)
```

```python
df_t=pd.DataFrame({'score':[a,b,c]})
df_t['rank']=df_t['score'].rank().astype(int)
df_t['weight']=df_t['rank']/df_t['rank'].sum()
df_t
```

```python
from sklearn.metrics import r2_score 
weight0=prediction1*float(df_t.loc[[0],['weight']].values)
weight1=prediction2*float(df_t.loc[[1],['weight']].values)
weight2=prediction3*float(df_t.loc[[2],['weight']].values)
total=weight0+weight1+weight2
r2_score(val_y,total)
```
