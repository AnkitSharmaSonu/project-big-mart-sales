# cross fold validation

```python
from sklearn.preprocessing import StandardScaler 
ss=StandardScaler()
x=ss.fit_transform(x)
```

```python
from sklearn.neighbors import KNeighborsClassifier as knn
from sklearn.model_selection import cross_val_score as cc
z=cc(knn(n_neighbors=3),train_x,train_y,cv=10)
z
```

```python
def val(k):
    me=[]
    sd=[]
    for i in k:
        z=cc(knn(n_neighbors=i),train_x,train_y,cv=10)
        me.append(z.mean())
        sd.append(z.std())
    return me,sd
```

```python
x= range(1,50)
mean,std = val(x)
std
```

```python
plt.plot(x[10:20],mean[10:20], color = 'green', label = 'mean' )
plt.xlabel('n_neighbors')
plt.ylabel('Mean Score')
plt.title('Mean Validation score')
```

```python
plt.plot(k[10:20],std[10:20],color='green',label='mean')
plt.xlabel('n_neighbors')
plt.ylabel('standard deviation')
plt.title('standard')
```

```python
k=knn(n_neighbors=10)
k.fit(train_x,train_y)
score1=k.score(train_x,train_y)
score2=k.score(test_x,test_y)
print(score1,score2)

```
