```python
df['Item_Identifier'].value_counts()
```

```python
tmp=df['Item_Identifier'].value_counts()
tmp.head()
```

```python
df['Item_Identifier_count']=df['Item_Identifier'].apply(lambda x: tmp[x])
df[['Item_Identifier','Item_Identifier_count']].head()
```

```python
for i in range(len(df['Item_Identifier'])):
    if df['Item_Identifier_count'][i]<4:
        df['Item_Identifier'][i]='others'
df['Item_Identifier'].value_counts()
```

```python
df1=pd.read_csv('stroke prediction.csv')
df1.head()
```

```python
bins=[0,12,20,30,50,70]
grp=['child','teenage','young adult','mature adult','senior']
df1['Age_group']=pd.cut(df1['age'],bins,labels=grp)
df1[['age','Age_group']]
```

```python
df1['Missing_smoking']=df1['smoking_status'].isnull()*1
df1[['smoking_status','Missing_smoking']]
```
