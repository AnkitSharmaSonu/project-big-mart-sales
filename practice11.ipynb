```python
import matplotlib.pyplot as plt
import numpy as np
x1=[1,2,3,4]
y1=[5,6,7,8]
x=np.arange(0,1,0.1)
y=1+np.cos(2*np.pi*x)
plt.subplot(1,2,1)
plt.plot(x,y)
plt.grid()
plt.xlabel('this is x axis')
plt.ylabel('this is y-axis')
plt.title('this is first practice graph')
plt.subplot(1,2,2)
plt.plot(x1,y1)
plt.grid()
plt.show()
```
