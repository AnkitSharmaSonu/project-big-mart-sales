```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('train_cleaned.csv')
data.head()
```

```python
x=data.drop(['Item_Outlet_Sales'],axis=1)
y=data['Item_Outlet_Sales']
```

```python
from sklearn.model_selection import train_test_split as tt
train_x,test_x,train_y,test_y=tt(x,y,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression as LR
from sklearn.metrics import mean_absolute_error as mae
```

```python
l=LR()
l.fit(train_x,train_y)
train_fit=l.predict(train_x)
```

```python
error=mae(train_fit,train_y)
error
```

```python
test_predict=l.predict(test_x)
error1=mae(test_predict,test_y)
error1
```

```python
plt.figure(figsize=(8,6),facecolor='w',edgecolor='b',dpi=120)
x = range(len(train_x.columns))
y=l.coef_
plt.bar(x,y)
```

```python
res = pd.DataFrame({
    'fitted values' : test_y,
    'predicted values' : test_predict
})
res['residuals']=res['fitted values']-res['predicted values']
```

```python
plt.figure(figsize=(8,6),facecolor='b',edgecolor='w',dpi=120)
k=range(0,2131)
f=[0 for i in k]
plt.scatter(k,res['residuals'])
plt.plot( f, k , color = 'red', label = 'regression line' )
plt.xlabel('fitted points ')
plt.ylabel('residuals')
plt.title('Residual plot')
plt.ylim(-4000, 4000)
```
