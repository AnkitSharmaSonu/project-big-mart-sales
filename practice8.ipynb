```python
import matplotlib.pyplot as plt
l=[1,2,3,4]
l1=[5,6,7,8]
l2=[9,10,11,12]
l3=[13,14,15,16]
plt.sublpot(2,1,1)
plt.plot(l,l1)
plt.subplot(2,1,2)
plt.plot(l2,l3)
plt.show()
```

```python

plt.bar(l,l1,width=0.8,color=['red','green'],tick_label=['one','two','three','four'])

plt.show()
```

```python
import matplotlib.pyplot as plt
l=[10,20,30,40,50,60,70,80,90]
bins=5
range=(1,100)
plt.hist(l,bins,range,color='red',rwidth=0.5)
plt.show()
```

```python
range=(10,30)
age=[20,30,100,22,33,23,15,38,42,43,26,16,30]
bins=10
plt.hist(age,bins,range,color='red',rwidth=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
l=[1,2,3,4]
l1=[5,6,7,8]
plt.scatter(l,l1,label='ankit',marker='*',s=30,color='red')
plt.legend()
plt.show()
```

```python
ea=['one','two','three','four']
plt.pie(l,labels=ea,colors=['r','g','b'],shadow=True,startangle=90,explode=(0,0.1,0,0.1),autopct='%1.2ff')
plt.legend()
plt.show()
```
