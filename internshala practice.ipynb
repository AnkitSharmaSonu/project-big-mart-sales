```python
for i in range(0,len(df3['N_BATHROOM'])):
    if pd.isnull(df3['N_BATHROOM'][i])==True:
        if pd.isnull(df3['N_BATHROOM'][i])==1.0:
            
            
        else:
            df3['N_BATHROOM']=df3['N_BATHROOM'].mode()[0]
```

```python
df3.isnull().sum()
```

```python
temp=(df3['N_ROOM']+df3['N_BATHROOM']+df3['N_BEDROOM'])/3
```

```python
pd.concat([df3['N_ROOM'],df3['N_BATHROOM'],df3['N_BEDROOM'],temp],axis=1).head()
```

```python
df3.loc[df3['QS_OVERALL'].isnull()==True].shape
```

```python
def fill_na(x):
    return ((x['QS_ROOMS'] + x['QS_BATHROOM'] + x['QS_BEDROOM'])/3)
```

```python
df3['QS_OVERALL']=df3.apply(lambda x: fill_na(x) if pd.isnull(x['QS_OVERALL']) else x['QS_OVERALL'] ,axis=1)
```

# Regression benchmark 


```python
df4=pd.read_csv('train_bm.csv')
df4.head()
```

```python
df4.head()
```

```python
df4.isnull().sum()
```

```python
from sklearn.utils import shuffle
df4=shuffle(df4,random_state=42)
div=df4[0:3].shape[0]//3

train=df4.loc[0:3*div+1,:]
test=df4.loc[3*div+1:,:]
```

```python
train.head()
```

```python
test.head()
```

```python
test['simple_mean'] = train['Item_Outlet_Sales'].mean()
```

```python
test['simple_mean']
```

```python
from sklearn.metrics import mean_absolute_error as MAE

simple_mean_error = MAE(test['Item_Outlet_Sales'] , test['simple_mean'])
simple_mean_error
```
